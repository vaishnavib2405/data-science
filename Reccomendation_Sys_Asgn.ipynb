```python
df3=df2.pivot_table(index='UserID',columns='Title',values='Rating').reset_index(drop=True)
df3
```

```python
df3.index=df2['UserID'].unique()
df3
```

```python
df3.fillna(0,inplace=True)
df3
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1-pairwise_distances(df3.values,metric='cosine')
user_sim
```

```python
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
user_sim2.index=df2['UserID'].unique()
user_sim2.columns=df2['UserID'].unique()
user_sim2
```
