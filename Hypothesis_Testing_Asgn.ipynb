# Hypothesis Testing exercise 1 solution

```python
# Null hypothesis h0 = Diameters of the cutlets of the two units are similar
# Alternate hypothesis ha = Diameters of the cutlets of the two units are dissimilar
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("Cutlets.csv")
```

```python
df.head()
```

```python
#let mu1 be the mean of cutlets of unit A and mu2 be the mean of cutlets of unit B
#Hence h0 : mu1 - mu2 = 0 & ha: mu1 - mu2 != 0
unit_A = df['Unit A']
unit_B = df['Unit B']
```

```python
#Since this is a sample we have to use t test
stats.ttest_ind(unit_A , unit_B)
```

```python
#Here significence level is 5% so alpha is 0.05
#pvalue ~ 0.4722 which is greater than alpha which means that the diameters of the cutlets of the two units are not significantly different and hence we accept h0
```

# Hypothesis Testing Exercise 2 solution

```python
#Here we have 4 different types laboratory so let their respective means be mu1,mu2,mu3,mu4
#Null hypothesis h0 = all the means are equal i.e mu1=mu2=mu3=mu4.
#Alternative hypothesis ha = atleast one of the mean is different.
from google.colab import files
uploaded = files.upload()
```

```python
df2 = pd.read_csv('LabTAT.csv')
```

```python
df2.head()
```

```python
df2.shape
```

```python
#Since we have to find if the means of 4 columns are equal or not, we have to use f test or anova test
stats.f_oneway(df2.iloc[:,0], df2.iloc[:,1],df2.iloc[:,2],df2.iloc[:,3])
```

```python
#Here p value is almost nearly zero and is thus less than alpha or significance level i.e 5% or 0.05. This means that we reject h0 and accept ha
#Hence, there is indeed a difference in the average TAT values between the 4 laboratories
```

```python
df3.head()
```

```python
#Here, h0 = Proportions are equal and ha = Proportions are not equal
```

```python
data = df3.iloc[:,1:5]
```

```python
data
```

```python
#Since there are counts of males and females across different regions, we use chi square to find relation between them.
```

```python
stats.chi2_contingency(data)
```

```python
#Here the p value is 0.6603 which is greater than 0.05 i.e alpha (significance level). Hence proportions are not equal.
```
